### **Example-1**

```python
import matplotlib.pyplot as plt

# step-1: prepare data
x=[1, 2, 3, 4, 5]
y=[30, 35, 40, 45, 50]

# step-2: prepare plot
plt.plot(x,y)

# step-3: customize
plt.title("Sample Line Plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")

#step-4: show the plot
plt.show()
```

### **Example-2**

```python
import pandas as pd
import matplotlib.pyplot as plt

# Sample dataset
data = { 'Day':[1, 2, 3, 4, 5],
        'Price':[101, 106, 99, 109, 112]}

df = pd.DataFrame(data)

# customize
plt.plot(df['Day'],df['Price'],marker='o')
plt.title("Stock Price Trend")
plt.xlabel("day")
plt.ylabel("price")
plt.show()
```

### **Example-1**

```python
import matplotlib.pyplot as plt

products = ['A', 'B', 'C', 'D']
sales = [110, 87, 160, 135]

plt.bar(products, sales, color='pink',edgecolor='black')
plt.title("Sales by Products")
plt.xlabel("Products")
plt.ylabel("Sales")

# show the barchart
plt.show
```

```python
import numpy as np

returns = np.random.normal(0,1,1000)
plt.hist(returns,bins=30,color='skyblue',edgecolor='black')
plt.title("Stock returns Distribution")
plt.xlabel("Return")
plt.ylabel("Frequency")
plt.show()
```

```python
import matplotlib.pyplot as plt

volume = [100, 130, 170, 119, 220]
prices = [100, 98, 102, 107, 105]

plt.scatter(volume, prices, color='red')
plt.title("Volume vs prices")
plt.xlabel("Volume")
plt.ylabel("Price")
plt.show()
```

### **Example-2**

```python
import matplotlib.pyplot as plt

study_hours = [2,3,4,5,6,7,8,9,10]
exam_scores = [50, 55, 60, 65, 70, 75, 80, 85, 90]

plt.figure(figsize=(7,5))
plt.scatter(study_hours,exam_scores,color="green",marker="*", s=120,alpha=0.8)

plt.title("Study Hours vs Exam Scores")
plt.xlabel("Hours of study")
plt.ylabel("Exam score (%)")
plt.grid(True)

plt.show()
```

```python
import matplotlib.pyplot as plt

# Sample exam scores for 3 subjects
math_scores = [65,70,72,80,85,90,95,100,55,60]
science_scores = [50,60,77,90,100,83,85,65,70,80]
english_scores = [55,60,65,70,72,74,78,80,82,85]

# combine into a list
data = [math_scores, science_scores, english_scores]

# box plot
plt.figure(figsize=(7,5))
plt.boxplot(data,patch_artist=True,labels=["Math","Science","English"])

plt.title("Exam scores distribution by subject")
plt.ylabel("Scores")
plt.grid(True)

plt.show()
```
